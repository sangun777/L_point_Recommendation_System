# A01

```python
type_predict=pd.read_csv('./type_score.csv')
type_label=pd.read_csv('./type_test.csv')
brand_predict=pd.read_csv('./brand_type_best1_score.csv')
brand_label=pd.read_csv('./brand_type_best1_test_label.csv') #418 -> 432
```

```python
type_label
```

```python
brand_label
```

```python
brand_label=brand_label.iloc[:346,:]
brand_predict=brand_predict.iloc[:346,:]
```

```python
type_predict.drop('Unnamed: 0',axis=1,inplace=True)
brand_predict.drop('Unnamed: 0',axis=1,inplace=True)
type_label.drop('Unnamed: 0',axis=1,inplace=True)
brand_label.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
type_predict
```

```python
brand_predict2=brand_predict.T
brand_label2=brand_label.T
type_predict2=type_predict.T
type_label2=type_label.T
```

```python
type_label2
```

```python
brand_count=[]
for i in range(len(brand_predict)):
    ss=brand_label2[i].loc[brand_label2[i]>0].index
    dd=brand_predict2[i].reset_index()
    dd=dd.sort_values(i,ascending=False).iloc[:5,0]
    ff=[]
    for j in range(len(ss)):
        if any(dd.str.contains(ss[j])):
            ff.append('1')
        else:
            ff.append('0')
    if any('1' in k for k in ff):
        brand_count.append(1)
    else:
        brand_count.append(0)
    print(i)
```

```python
type_count=[]
for i in range(len(type_predict)):
    ss=type_label2[i].loc[type_label2[i]>0].index
    dd=type_predict2[i].reset_index()
    dd=dd.sort_values(i,ascending=False).iloc[:5,0]
    ff=[]
    for j in range(len(ss)):
        if any(dd.str.contains(ss[j])):
            ff.append('1')
        else:
            ff.append('0')
    if any('1' in k for k in ff):
        type_count.append(1)
    else:
        type_count.append(0)
    print(i)
```
